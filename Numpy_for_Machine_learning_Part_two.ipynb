```python
import numpy as np
```

```python
arr = np.random.randint(1,100,size = (3,5))
print(arr)
```

```python
arr_2 = arr.reshape(5,3)
print(arr_2)
```

```python
flat_as_row = arr_2.flatten() #row wise flat
print(flat_as_row)
```

```python
flat_as_col = np.ravel(arr_2,order = 'F') #column wise flat
print(flat_as_col)
```

```python
a = np.random.randint(1,10,size=(2,4))
b = np.random.randint(10,20,size=(2,4))
print(a)
print(b)
```

```python
#array split
a = np.random.randint(1,10,size = (10,))
print(a)
b = np.array_split(a,3)
print(b)
```

```python
#for equal division, so that each slice get the same number of entries
a = np.random.randint(1,10,size = (10,))
b = np.split(a,3) #its not possible, as we can't divide 10 equally for 3 slice
#so it will show error
print(b)
```

```python
x = np.array([ 10 , 8 , 30 , 100 ])
y = np.array([ 2 , 3 ,4 ,5 ])
add = x + y
sub = x-y
mul = x*y
div = x/y
print(add)
print(sub)
print(mul)
print(div)
```

#mathematics and trigonometry

```python
x = np.array([10,20,30,40,45,90])
sin_x = np.sin(x) #converts from radian not from degree
print(sin_x)
cos_x = np.cos(x)
print(cos_x)
tan_x = np.tan(x)
print(tan_x)
```

#Degree to Radian Conversion

```python
deg = np.array([10,20,30,40,50])
rad = np.deg2rad(deg)
print(rad)
```

```python
deg = np.rad2deg(rad)
print(deg)
```

```python
x = np.array([ 10 , 8 , 16 , 100 ])
base_10_log = np.log10(x)
base_2_log = np.log2(x)
print(base_10_log)
print(base_2_log)
```

```python
x = np.array([ 10 , 8 , 16 , 100 ])
sums = np.sum(x)
cum_sums = np.cumsum(x)
print(sums)
print(cum_sums)
```

#Logical Function

```python
x = np.array([ 10 , 8 , 16 , 100 ])
y = np.array([ 2 , 3 ,16 ,5 ])
greater_than = x>y
print(greater_than)
less_than = x<y
print(less_than)
equals = (x==y)
print(equals)
```

```python
print(np.all(greater_than))
```

```python
x = np.array([ 10 , 8 , 16 , 100 ])
print(x)
y = x.copy()
print(y)
y.sort()
print(y)
```

```python
mat = np.array([[10,3,5],[8,4,9]])
hor_sort = np.sort(mat,axis = 1)
ver_sort = np.sort(mat,axis = 0)
print(mat)
print(hor_sort)
print(ver_sort)
```

#value count in ndarray

```python
a = np.random.randint(1 ,100 , size =(100,))
print(a)
```

```python
val_greater_than_50 = np.count_nonzero(a>50)
print(val_greater_than_50)
```

```python
unique_val,count = np.unique(a,return_counts = True)
print(unique_val)
print(count)
```
