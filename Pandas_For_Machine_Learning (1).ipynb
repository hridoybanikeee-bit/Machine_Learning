#Filtering data based on Conditions

```python
not_started = df.loc[df['CompletionStatus']== 'Not Started']
print(not_started)
```

```python
completed = df.loc[df['CompletionStatus']== 'Completed']
print(completed)
```

```python
more_than_90 = df.loc[(df['Data Structure Marks'] > 90) & (df['CompletionStatus'] == 'Completed')]
print(more_than_90)
```
