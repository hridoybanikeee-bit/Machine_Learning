Number of Words

```python
from functools import reduce
with open("/content/sample_data/bilotaa.txt","r") as file:
  lst = file.readlines()
  words = list(map(lambda x: x.split(), lst))
  print(words)
  word_count = list(map(lambda y: len(y), words))
  print(word_count)
  overall_count_words = reduce(lambda x,y: x+y, word_count)
  print(overall_count_words)
```

Number of Characters

```python
from functools import reduce
with open("/content/sample_data/bilotaa.txt","r") as file:
  lst = file.readlines()
  cancel_new_line = list(map(lambda x: x.strip(), lst))
  remove_space = list(map(lambda x: x.replace(" ",""), cancel_new_line))
  each_character_count = list(map(lambda x: len(x), remove_space))
  overall_count_characters = reduce(lambda x,y: x+y, each_character_count)
  print(overall_count_characters)
```

```python
with open("/content/sample_data/output.txt","w") as file:
  file.write("We will show some results here\n")
  file.write(f"Total number of lines {len(lst)}\n")
  file.write(f"Total number of lines {overall_count_words}\n")
  file.write(f"Total number of lines {overall_count_characters}\n")
```
