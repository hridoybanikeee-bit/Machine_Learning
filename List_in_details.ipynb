**Loop through a list**

```python
list5 = ["apple","cherry","jack fruit","Banana","Jam"]
for i in list5:
  print(i)
```

```python
for i in range(len(list5)):
  print(list5[i])
```

```python
i = 0;
while(i<len(list5)):
  print(list5[i])
  i = i+1
```

```python
[print(i) for i in list5]
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
new = []
for x in fruits:
  if 'a' in x:
    new.append(x)

print(new)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
mylist = [x for x in fruits if 'a' in x]
print(mylist)
```

```python
new_list = [x for x in fruits]
print(new_list)
```

```python
another_list = [x for x in range(15) if x%2 == 0]
print(another_list)
```

**Sorting List**

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango","bin"]
fruits.sort()
print(fruits)
```

```python
num = [10,3,4,34,4,6,4,54,64]
num.sort()
print(num)
```

```python
num.sort(reverse = True)
print(num)
```

**Join List**

```python
lista = ['a','b','c','d','e']
listb = ['e', 'f', 'g', 'h']
listc = lista + listb
print(listc)
```

```python
lista = ['a','b','c','d','e']
listb = ['e', 'f', 'g', 'h']
for x in listb:
  lista.append(x)
print(lista)
```

```python
lista = ['a','b','c','d','e']
listb = ['e', 'f', 'g', 'h']
lista.extend(listb)
print(lista)
print(listb)
```
