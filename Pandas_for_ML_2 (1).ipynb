```python
#to find out if New exists in the city column at any place
df.loc[df['City'].str.contains("New")]
```

```python
#to find out if it captures with small letter, which is not possible obviously
df.loc[df['City'].str.contains("new")]
```

```python
#now by using case attribute, we are telling the dataset about it doesn't matter whether its written on capital or small letter
df.loc[df['City'].str.contains("new", case=False)]
```

```python
#to find out if Los exists in the beggining of the string
df.loc[df['City'].str.contains("^Los")]
```

```python
#to find out if rk exists in the end of the string
df.loc[df['City'].str.contains(f"rk$")]
```

```python
#if name exists with vowel
df.loc[df['Name'].str.contains(r"^[AEIOU]")]
```

```python
#string Contains New or Los
df.loc[df['City'].str.contains(f"New | Los")]
```

#group by functions in pandas

```python
df = pd.read_csv("/content/student_completed_data.csv")
display(df)
```

```python
df.loc[df['FullName'] == 'Afsana Mimi', 'Total Marks'] = np.nan
df.loc[df['FullName'] == 'Sadia Chowdhury', 'Total Marks'] = np.nan
df.loc[df['FullName'] == 'Alif Rahman', 'Total Marks'] = np.nan
display(df)
```

```python
instructor_group = df.groupby('Instructor')
for instructor,values in instructor_group:
  print(instructor)
  print(values)
```

```python
instructor_group.describe()
```

```python
mean_marks = instructor_group['Total Marks'].mean()
display(mean_marks)
```

```python
df['Total Marks'] = df['Total Marks'].fillna(df['Instructor'].map(mean_marks))
display(df)
```
