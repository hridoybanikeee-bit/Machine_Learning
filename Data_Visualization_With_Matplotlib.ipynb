```python
df = pd.read_csv('/content/enrollment_data.csv')
df
```

```python
plt.plot(df['Year'],df['Digital Marketing'])
plt.plot(df['Year'],df['Programming'])
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title("Student Enrollment Data")
plt.show()
```
