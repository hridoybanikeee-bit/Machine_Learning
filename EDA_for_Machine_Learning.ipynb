#Question 10 (10 Marks)  
Handle missing values in the Titanic dataset by:  
Imputing missing Age values using SimpleImputer with mean strategy  
Imputing missing Embarked values using most frequent strategy  
Imputing missing Cabin values using constant strategy with "Missing"  
Finally, check whether any missing values remain.  


```python
from sklearn.impute import SimpleImputer
imputer_mean = SimpleImputer(missing_values = np.nan, strategy = "mean")

imputer_mean.fit(X_train[['Age']])
X_train['Age'] = imputer_mean.transform(X_train[['Age']]).ravel()
X_test['Age'] = imputer_mean.transform(X_test[['Age']]).ravel()

imputer_most_frequent = SimpleImputer(missing_values = np.nan, strategy = "most_frequent")
imputer_most_frequent.fit(X_train[['Embarked']])
X_train['Embarked'] = imputer_most_frequent.transform(X_train[['Embarked']]).ravel()
X_test['Embarked'] = imputer_most_frequent.transform(X_test[['Embarked']]).ravel()


imputer_with_constant = SimpleImputer(missing_values = np.nan, strategy = "constant", fill_value = "Missing")
imputer_with_constant.fit(X_train[['Cabin']])
X_train['Cabin'] = imputer_with_constant.transform(X_train[['Cabin']]).ravel()
X_test['Cabin'] = imputer_with_constant.transform(X_test[['Cabin']]).ravel()
```
