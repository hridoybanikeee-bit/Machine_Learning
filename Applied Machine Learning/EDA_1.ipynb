#Task 7: <br>
Let's visualize the gender distribution among the surveyed penguins. <br>
• Extract the value counts for the sex column.<br>
• Create a pie chart using Matplotlib showing the male vs. female distribution.<br>
• Ensure the pie chart includes percentage labels with one decimal point (example 50.5%).
Add a title to your chart.

```python
counts = df['sex'].value_counts()
labels = counts.index
```

```python
plt.pie(counts, labels = labels, autopct='%1.1f%%')
plt.title("Pie Charts of Gender Distribution")
```
