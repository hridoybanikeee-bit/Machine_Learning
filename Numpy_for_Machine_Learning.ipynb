**NUMPY** uses *SIMD (Single Instruction Multiple Data)* method. As, list uses single instruction one by one, it is slow and not efficient. \n
On the other hand numpy uses parallel computation method. That's why it is must faster.

```python
arr = [1,2,3]
print(arr*5)
arr2 = [x*5 for x in arr]
print(arr2)
```

```python
arr1 = np.array([1,2,3])
arr2 = arr1 * 5
print(arr2)
```

**Dimension of array or matrix**

```python
#Nd array (N dimensional array)

#1D array

arr1 = np.array([1,2,3,4,5,6])
print(type(arr1))
print(arr1.ndim)
```

```python
#2D array
arr2 = np.array([[1,2,3],
                 [4,5,6]])
print(arr2)
print(arr2.ndim)
```

```python
#3D array
arr3 = np.array([
    [[1,2,3],
    [4,5,6]],
    [[1,2,3],
     [3,4,5]],

    [[1,2,3],
     [3,4,5]]

])
print(arr3)
print(arr3.ndim)
```

Advance indexing of Numpy

```python
#view mode
arr1 = np.array([1,2,3,4,5,6,7])
arr_one = arr1[0:4]
print(arr_one)
arr_one[2] = 100
print(arr_one)
print(arr1)
```

```python
#copy mode
arr1 = np.array([1,2,3,4,5,6,7])
arr_one = arr1[0:4].copy()
print(arr_one)
arr_one[2] = 100
print(arr_one)
print(arr1)
```

1. You have a list of temperatures recorded in Celsius. Your task is to:
Create a NumPy array from the given list.
Convert all temperatures from Celsius to Fahrenheit. The formula is F = C* 1.8 + 32.
Print the new Fahrenheit array.


```python
temp = [10, 45, 54, 100, 36, 4, 0]
arr = np.array(temp)
faren_arr = arr* 1.8 +32
print(faren_arr)
```

You have a Python list containing the lap times (in seconds) for an athlete.

Task:

Convert the list into a NumPy array.
Print the array's shape, size, and data type.
The athlete wants to see their times in minutes. Create a new array by dividing all the lap times by 60.
Print the new array of lap times in minutes


```python
lap_times = [62.5, 64.2, 30, 65.1, 66.4, 62.9, 61.7, 63.5, 64.0, 60.8]
arr = np.array(lap_times)
print(arr.shape)
print(arr.size)
print(arr.dtype)
min_arr = arr/60
print(min_arr)
```
