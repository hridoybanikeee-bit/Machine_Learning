**Basic ATM project**

```python
class Atm:
  def __init__(self):
    self.pin = ''
    self.balance = 0
    self.menu()

  def menu(self):
    user_input = input(
        """
        Welcome to my Bank, which I've created! Ha Ha
        1. Press 1 to create pin
        2. Press 2 to change pin
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. Anything else to exit
        """)
    if user_input == '1':
      self.create_pin()
    elif user_input == '2':
      self.change_pin()
    elif user_input == '3':
      self.check_balance()
    elif user_input == '4':
      self.withdraw()
    else:
      exit()

  def create_pin(self):
    user_pin = input("Enter your pin: ")
    self.pin = user_pin
    balanceee = int(input("Enter balance: "))
    self.balance = balanceee
    print("PIN Change successful!")
    self.menu()
  def change_pin(self):
    old_pin = input('Enter your old pin to verify: ')
    if old_pin != self.pin:
      print("Authorized Denied! Please type the correct PIN")
      self.menu()
    else:
      new_pin = input("Please enter a new pin: ")
      self.pin = new_pin
      self.menu()
  def check_balance(self):
    pin = input("Enter your PIN: ")
    if pin == self.pin:
      print(f"Your account balance is {self.balance}")
      self.menu()
    else:
      print("Authorized Denied! Please type the correct PIN")
  def withdraw(self):
     pin = input("Enter your PIN: ")
     if pin == self.pin:
      amount = int(input("Enter your withdraw amount: "))
      if amount > self.balance:
        print("Sorry! Insufficient amount")
        self.menu()
      else:
        print("Successfully authorized transection")
        remaining_balance = self.balance - amount
        self.balance = remaining_balance
        print(f'Remaining Balance: {remaining_balance}')
        self.menu()
     else:
        print("Authorized Denied! Please type the correct PIN")
        self.menu()



```

**Fraction Class**

```python

class Fraction:
  def __init__(self, x, y):
    self.num = x
    self.den = y
  def __str__(self):
    return '{}/{}'.format(self.num, self.den)
  def __add__(self,other):
    new_num = self.num*other.den + self.den*other.num
    new_denum = self.den * other.den
    return '{}/{}'.format(new_num, new_denum)
  def __sub__(self,other):
    new_num = self.num*other.den - self.den*other.num
    new_denum = self.den * other.den
    return '{}/{}'.format(new_num, new_denum)
  def __mul__(self,other):
    new_num = self.num * other.num
    new_denum = self.den * other.den
    return '{}/{}'.format(new_num, new_denum)
  def __truediv__(self,other):
    new_num = self.num * other.den
    new_denum = self.den * other.num
    return '{}/{}'.format(new_num, new_denum)

fr1 = Fraction(3,5)
fr2 = Fraction(4,6)
print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)

```
