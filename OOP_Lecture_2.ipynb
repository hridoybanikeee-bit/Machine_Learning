Write OOP classes to handle the following scenarios:
• A user can create and view 2D coordinates
• A user can find out the distance between 2 coordinates
• A user can find find the distance of a coordinate from origin
• A user can check if a point lies on a given line






```python
class Point:
  def __init__(self,x_cod,y_cod):
    self.x = x_cod
    self.y = y_cod
  def str(self):
    return '({},{})'.format(self.x,self.y)
  def distance(self,other):
    dis = ((self.x - other.x)**2 + (self.y - other.y)**2)**0.5
    return dis
  def distance_origin(self):
    dist = ((self.x)**2 + (self.y)**2)**0.5
    return dist

```

```python
class Line:
  #ax + by + c = 0
  def __init__(self,a,b,c):
    self.a = a
    self.b = b
    self.c = c
  def str(self):
    return '{}x + {}y + {} = 0'.format(self.a, self.b, self.c)
  def point_on_line(line,point):
    if(line.a * point.x + line.b*point.y + line.c == 0):
      return 'Lies on the line'
    else:
      return 'Does not lies on the line'

```

```python
line = Line(2, -1, -1)   # 2x - y - 1 = 0
p1 = Point(1, 1)
line.point_on_line(p1)
```

```python
p1 = Point(5,7)
p2 = Point(3,5)
print(p1.distance(p2))
print(p1.distance_origin())
print(p2.distance_origin())
```

```python
l = Line(2,3,5)
l.point_on_line(p1)
```

How Objects access attributes


```python
class Person:
  def __init__(self,name,country):
    self.name = name
    self.country = country
  def greet(self):
    if self.country == "Bangladesh":
      print(f"Papi Bilota ka Swagat: {self.name}")
    else:
      print(f"Bhaag saale: {self.name}")
```

```python
akash = Person("Akash","Bangladesh")
niloy = Person("Niloy","India")
akash.greet()
niloy.greet()
```

```python
class Atm:
  def __init__(self):
    self.pin = ''
    self.__balance = 0
    self.menu()

  def get_bal(self):
    return self.__balance

  def set_bal(self,new_val):
    if type(new_val) == int:
      self.__balance = new_val
    else:
      print("Nope! It won't work")

  def menu(self):
    user_input = input(
        """
        Welcome to my Bank, which I've created! Ha Ha
        1. Press 1 to create pin
        2. Press 2 to change pin
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. Anything else to exit
        """)
    if user_input == '1':
      self.create_pin()
    elif user_input == '2':
      self.change_pin()
    elif user_input == '3':
      self.check_balance()
    elif user_input == '4':
      self.withdraw()
    else:
      exit()

  def create_pin(self):
    user_pin = input("Enter your pin: ")
    self.pin = user_pin
    balanceee = int(input("Enter balance: "))
    self.__balance = balanceee
    print("PIN Change successful!")
    self.menu()
  def change_pin(self):
    old_pin = input('Enter your old pin to verify: ')
    if old_pin != self.pin:
      print("Authorized Denied! Please type the correct PIN")
      self.menu()
    else:
      new_pin = input("Please enter a new pin: ")
      self.pin = new_pin
      self.menu()
  def check_balance(self):
    pin = input("Enter your PIN: ")
    if pin == self.pin:
      print(f"Your account balance is {self.__balance}")
      self.menu()
    else:
      print("Authorized Denied! Please type the correct PIN")
  def withdraw(self):
     pin = input("Enter your PIN: ")
     if pin == self.pin:
      amount = int(input("Enter your withdraw amount: "))
      if amount > self.__balance:
        print("Sorry! Insufficient amount")
        self.menu()
      else:
        print("Successfully authorized transection")
        remaining_balance = self.__balance - amount
        self.__balance = remaining_balance
        print(f'Remaining Balance: {remaining_balance}')
        self.menu()
     else:
        print("Authorized Denied! Please type the correct PIN")
        self.menu()
```
